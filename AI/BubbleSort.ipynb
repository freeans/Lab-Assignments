```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Example usage
arr = [64, 25, 12, 22, 11]
print("Before sorting:", arr)

bubble_sort(arr)

print("After sorting: ",arr)
```

```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
    return arr   

arr=[12,1,23,5,21,4]

print("Before Sorting:",arr)

bubble_sort(arr)
print("After Sorting:",arr)
```
