```python
def toh(n,source,auxiliary,destination):
    if n==1:
        print(f"Move disk1 from {source} to {destination}")
        return
    toh(n-1,source,destination,auxiliary)
    print(f"Move disk{n} from {source} to {destination}")
    toh(n-1,auxiliary,source,destination)


if __name__=="__main__":
    num_disk=int(input("Enter The Number Of Disk"))
    print("The Sequence Of Disk Moved Is:")
    toh(num_disk,'A','B','C')
```

```python
def toh(n,source,auxilary,destination):
    if n==1:
        print(f"Move Disk 1 from {source} To {destination}")
        return
    toh(n-1,source,destination,auxilary)
    print(f"Move Disk {n} from {source} to {destination}")
    toh(n-1,auxilary,source,destination)

if __name__=="__main__":
    num=int(input("Enter The Number Of Disks:"))
    print("The Sequence Of Disk Moved Is:")
    toh(num,'A','B','C')
```
