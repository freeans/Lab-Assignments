## Actual Program Starts Here

```python
import re
import pandas as pd
```

```python
#download twitter sentiments.csv
data = pd.read_csv("twitter_sentiments.csv", names=['id', 'loc', 'label', 'text'])
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
data['text'] = data['text'].astype(str)
```

```python
data.dtypes
```

```python
#clening of text
def clean_text(text):
  text = text.lower() #lowercase
  text = re.sub(r"[^a-zA-Z]+", " ", text) #remove non-aplhanumric chars
  return text
```

```python
clean_text("hello @3423432 friends, im fine;;;;")
```

```python
data['text'] = data['text'].apply(clean_text)
```

```python
data
```

```python
comments = data['text'].tolist()
targets=data['label'].values
```

```python
pd.DataFrame(targets).value_counts()
```

```python
# Tokanization and padding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words = 5000)
tokenizer.fit_on_texts(comments)
sequences = tokenizer.texts_to_sequences(comments)
padded_sequences = pad_sequences(sequences, maxlen=200)
```

```python
padded_sequences.shape
```

```python
padded_sequences
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(targets)

#one hot encoding
from keras.utils import to_categorical
y_new = to_categorical(y)
```

```python
y_new
```

```python
#train test split/ cross validation
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(padded_sequences, y_new, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Train Model

```python
# model definition (cutomized architecture)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

model = Sequential()
model.add(Embedding(5000, 128))
model.add(LSTM(64))
model.add(Dense(4, activation='softmax'))
#multi label output
```

```python
#compilation
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=32)
```

```python
## Output

new_comment = "I like him, saying this sentense is wrong thing..."
new_sequence = tokenizer.texts_to_sequences([clean_text(new_comment)])
padded_new_sequence = pad_sequences(new_sequence, maxlen=200)
prediction = model.predict(padded_new_sequence)[0]
```
